```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#Import Raw Data
agriculture_path = 'agriculture.csv'
emissions_path = 'emissions.csv'

emissions_data = pd.read_csv(emissions_path)
agriculture_data = pd.read_csv(agriculture_path)
```

```python
emissions_data.head()
# emissions_data.describe()
len(emissions_data)
```

```python
agriculture_data.head()
agriculture_data.describe()
```

```python
#keep only needed columns
agriculture_data = agriculture_data[['country'
                                     ,'year'
                                     ,'rice_yield']]

agriculture_data.head()
```

```python
#filter dataset to show only 1990 - 2020
agriculture_data = agriculture_data.loc[(agriculture_data['year'] >= 1990) 
                                        & (agriculture_data['year'] <= 2020)]


#check for null values
agriculture_data.describe()
```

```python
#remove null values from agriculture
agriculture_data = agriculture_data.dropna(axis=0)

agriculture_data.describe()
```

```python
#merge dataframes on country and year
merged_df = pd.merge(left=emissions_data
                     ,right=agriculture_data
                     ,left_on=['Area', 'Year']
                     ,right_on=['country', 'year']
                     ,how='left')

#drop redundant merge cols
merged_df = merged_df.drop(['country', 'year'], axis=1)

#rename country and year columns
merged_df = merged_df.rename(columns={'Area': 'Country'})

merged_df.describe()
```

```python
#count nulls for each col. display cols where null is zero
null_counts = merged_df.isnull().sum(axis = 0)
null_counts.loc[null_counts == 0 ]
```

```python
#rename merged dataframe 
#this is the final df ready for analysis
#drop na rows if needed
main_df = merged_df.reset_index(drop=True)
main_df
```
